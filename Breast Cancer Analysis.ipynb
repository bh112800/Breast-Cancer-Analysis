## Data Splitting

```python
# Splitting the data into train and test
from sklearn.model_selection import train_test_split

X = data[['radius_mean',  'perimeter_mean', 'area_mean', 'symmetry_mean', 'compactness_mean', 'concave points_mean']]
y = data.diagnosis

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30)
```

```python
# Scaling the data
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

## Model Implementation

```python
# Knn
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier()
param_knn = {'n_neighbors': range(1, 30)} 
grid_knn = GridSearchCV(knn, param_knn, cv = 5)
grid_knn.fit(X_train, y_train)

print('kNN: Best Paramater is', grid_knn.best_params_)
print('kNN: Best Mean Validation Score is', grid_knn.best_score_)
print("kNN: Accuarcy is", grid_knn.score(X_test, y_test))
```

```python
# Decision Tree
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier()
param_grid = {'max_depth':  range(1,10)         ,
              'min_samples_split':   range(2,11),
              'max_leaf_nodes':    range(2,11)   }


grid_tree = GridSearchCV(model, param_grid, cv=10)

grid_tree.fit(X_train, y_train) 

print("Decision Tree: Best Parameter is", grid_tree.best_params_)
print("Decision Tree: Best Score is", grid_tree.best_score_)
print("Decision Tree: Accuarcy is", grid_tree.score(X_test, y_test))
```

## Model Evaluation

```python
# kNN
from sklearn.metrics import precision_score, recall_score, accuracy_score, f1_score

knn_y_test_pred = grid_knn.predict(X_test)

print('kNN: Precision score:', precision_score(y_test, knn_y_test_pred))
print('kNN: Recall score:', recall_score(y_test, knn_y_test_pred))
print('kNN: Accuracy score:', accuracy_score(y_test, knn_y_test_pred))
print('kNN: F1 score:', f1_score(y_test, knn_y_test_pred))

# Precision-recall curve
from sklearn.metrics import precision_recall_curve

knn_prob = grid_knn.predict_proba(X_test)[:, 1]

knn_precision, knn_recall, knn_threshold = precision_recall_curve(y_test, knn_prob) # 3 outputs

# Plot
plt.plot(knn_recall, knn_precision)
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('kNN: Precision-Recall curve')

# AUC
from sklearn.metrics import roc_auc_score

print("kNN: AUC value is",  roc_auc_score(y_test, knn_prob) )
```

```python
# Decision Tree
from sklearn.metrics import precision_score, recall_score, accuracy_score, f1_score

tree_y_test_pred = grid_tree.predict(X_test)

print('Decision Tree: Precision score:', precision_score(y_test, tree_y_test_pred))
print('Decision Tree: Recall score:', recall_score(y_test, tree_y_test_pred))
print('Decision Tree: Accuracy score:', accuracy_score(y_test, tree_y_test_pred))
print('Decision Tree: F1 score:', f1_score(y_test, tree_y_test_pred))

# Precision-recall curve
from sklearn.metrics import precision_recall_curve

tree_prob = grid_tree.predict_proba(X_test)[:, 1]

tree_precision, tree_recall, tree_threshold = precision_recall_curve(y_test, tree_prob) # 3 outputs

# Plot
plt.plot(tree_recall, tree_precision)
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Decision Tree: Precision-Recall curve')

# AUC
from sklearn.metrics import roc_auc_score

print("Decision Tree: AUC value is",  roc_auc_score(y_test, tree_prob) )
```
